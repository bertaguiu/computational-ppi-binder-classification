# Processament dels resultats de Rosetta - IL-7Rα

Aquest notebook analitza les estructures generades per Rosetta Relax i calcula les mètriques energètiques globals i les mètriques específiques d'interfície per als 90 complexos del dataset IL-7Rα.

Els càlculs de Rosetta Relax proporcionen l'energia total del complex (score total en REU) però no permeten aïllar directament la contribució energètica de la interfície entre les dues cadenes. Per obtenir les mètriques d'interfície (ΔG, ΔSASA, packstat, ponts d'hidrogen, complementarietat de forma) és necessari aplicar una anàlisi addicional sobre les estructures optimitzades. Rosetta genera fitxers de sortida en format silent (.out) que contenen tota la informació estructural i energètica de les estructures generades en format binari, de manera que no cal reenviar els PDBs originals al clúster.

La funció analyseRosettaCalculation de la classe proteinModels llegeix els fitxers .out, reconstrueix les estructures mitjançant PyRosetta (executat dins l'entorn conda pp_pyro) i calcula les mètriques energètiques globals i les d'interfície per a la parella de cadenes A_B (target–binder). El resultat és un dataframe amb una fila per estructura i complex, que es guarda com a CSV per a l'agregació posterior.

```python
import sys
from pathlib import Path
from prepare_proteins import proteinModels
import pandas as pd

# Base del projecte
base = Path.cwd()

models_folder = base / "rosetta_relax_il7ra" / "input_models" # Carpeta on hi ha les estructures dels complexos .pdb

rosetta_folder = base / "rosetta_relax_il7ra" # Carpeta on hi ha els càlculs de Rosetta

sequence_models = proteinModels(str(models_folder))

rosetta_data = sequence_models.analyseRosettaCalculation(
    rosetta_folder=str(rosetta_folder),
    interface_chains="A_B",
    
    pyrosetta_env="pp_pyro",
    verbose=True,
    cpus=2
)

print("\n=== ROSETTA DATA ===")
print(rosetta_data.head())

output_csv = rosetta_folder / "rosetta_data_full.csv"
rosetta_data.to_csv(output_csv)

print(f"\nData guardada a: {output_csv}")
```
